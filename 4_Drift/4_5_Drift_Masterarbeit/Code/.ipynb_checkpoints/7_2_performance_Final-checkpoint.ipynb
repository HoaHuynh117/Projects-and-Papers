## Part 1: Statistical process control and window techniques

Sine, Hyperplane, MIXED, RandomRBF +
NaiveBayes +
ADWIN, DDM, EDDM, PageHinkley, KSWIN, HDDM_A, HDDM_W

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score
import time
from skmultiflow.data import SineGenerator, HyperplaneGenerator, STAGGERGenerator, MIXEDGenerator, SEAGenerator, RandomRBFGenerator
from skmultiflow.drift_detection import ADWIN, DDM, EDDM, PageHinkley, KSWIN, HDDM_A, HDDM_W
from skmultiflow.bayes import NaiveBayes

# Drift detection methods
detectors = {
    "ADWIN": ADWIN(),
    "DDM": DDM(),
    "EDDM": EDDM(),
    "HDDM_A": HDDM_A(),
    "HDDM_W": HDDM_W(),
    "KSWIN": KSWIN(),
    "PageHinkley": PageHinkley()
}

# Datasets
datasets = {
    "SineGenerator (Abrupt drift with a smooth, periodic change)": SineGenerator(has_noise=False, random_state=42),
    "HyperplaneGenerator (Gradual drift with a simple linear decision boundary)": HyperplaneGenerator(noise_percentage=0.1, n_drift_features=2, random_state=42),
    #"STAGGERGenerator (Abrupt Drift)": STAGGERGenerator(random_state=42),
    "MIXEDGenerator (Gradual drift with noise and multiple features)": MIXEDGenerator(random_state=42),
    #"SEAGenerator (Gradual drift in classification tasks)": SEAGenerator(random_state=42),
    "RandomRBFGenerator (Gradual drift with nonlinear transitions)": RandomRBFGenerator(model_random_state=42, sample_random_state=42)
}

# Experiment parameters
n_samples = 20000
chunk_size = 200

# Function to simulate stream and evaluate detectors with Naive Bayes
def evaluate_detectors_with_naive_bayes(dataset, detectors, n_samples, chunk_size):
    # Initialize results dictionary
    results = {name: {"accuracy": [], "false_alarms": [], "misses": [], "time": []} for name in detectors}
    results["Naive Bayes"] = {"accuracy": [], "false_alarms": [], "misses": [], "time": []}

    # Create stream
    stream_X, stream_y = dataset.next_sample(n_samples)
    true_drift_points = set(range(chunk_size, n_samples, chunk_size))

    # Initialize Naive Bayes classifier from skmultiflow
    nb_classifier = NaiveBayes()

    for detector_name, detector in detectors.items():
        print(f"Evaluating {detector_name}...")
        detector_start_time = time.time()
        detected_drifts = set()
        total_predictions = 0
        correct_predictions = 0

        # Naive Bayes tracking variables
        nb_start_time = time.time()
        nb_detected_drifts = set()
        nb_correct_predictions = 0

        # Track metrics for each chunk
        for i in range(0, n_samples, chunk_size):
            chunk_X = stream_X[i:i + chunk_size]
            chunk_y = stream_y[i:i + chunk_size]

            # Train Naive Bayes classifier on the current chunk
            nb_classifier.fit(chunk_X, chunk_y)
            nb_predictions = nb_classifier.predict(chunk_X)

            # Update Naive Bayes metrics
            nb_correct_predictions += accuracy_score(chunk_y, nb_predictions, normalize=False)

            # Simulate drift detection for each chunk
            for sample_X, sample_y in zip(chunk_X, chunk_y):
                detector.add_element(sample_y)
                if detector.detected_change():
                    detected_drifts.add(i)
                    break

            # Naive Bayes drift detection logic
            if i in true_drift_points:
                nb_detected_drifts.add(i)

            # Update total predictions and correct predictions for Naive Bayes
            total_predictions += len(chunk_y)
            correct_predictions += accuracy_score(chunk_y, nb_predictions, normalize=False)

            # Store results for each chunk
            results[detector_name]["accuracy"].append(correct_predictions / total_predictions if total_predictions > 0 else 0)
            false_alarms = len(detected_drifts - true_drift_points)
            misses = len(true_drift_points - detected_drifts)
            
            results[detector_name]["false_alarms"].append(false_alarms)
            results[detector_name]["misses"].append(misses)
            results[detector_name]["time"].append(time.time() - detector_start_time)

            # Collect Naive Bayes results
            results["Naive Bayes"]["accuracy"].append(nb_correct_predictions / total_predictions if total_predictions > 0 else 0)
            results["Naive Bayes"]["false_alarms"].append(len(nb_detected_drifts - true_drift_points))
            results["Naive Bayes"]["misses"].append(len(true_drift_points - nb_detected_drifts))
            results["Naive Bayes"]["time"].append(time.time() - nb_start_time)

    return results

# Run experiments with Naive Bayes
final_results_NB = {}
for dataset_name, dataset in datasets.items():
    print(f"\nDataset: {dataset_name}")
    final_results_NB[dataset_name] = evaluate_detectors_with_naive_bayes(dataset, detectors, n_samples, chunk_size)
    
def visualize_results_over_time(results):
    metrics = ["Accuracy", "False Alarms", "Misses", "Evaluation Time"]
    num_metrics = len(metrics)
    
    # Prepare data for visualization
    detector_names = [name for name in next(iter(results.values())).keys() if name != "Naive Bayes"]  # Exclude Naive Bayes
    datasets = list(results.keys())

    for dataset_name in datasets:
        #print(f"Visualizing results for {dataset_name}...")

        # Initialize subplots (2x2 grid for 4 metrics)
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        fig.suptitle(f"Results for {dataset_name}", fontsize=16)

        # Iterate through metrics and plot
        for i, metric in enumerate(metrics):
            metric_values = {detector_name: [] for detector_name in detector_names}
            time_points = list(range(0, n_samples // chunk_size))  # Time points based on chunks

            # Ensure the correct alignment of the metric values
            for detector_name in detector_names:
                if metric == "Accuracy":
                    metric_values[detector_name] = results[dataset_name][detector_name]["accuracy"]
                elif metric == "False Alarms":
                    metric_values[detector_name] = results[dataset_name][detector_name]["false_alarms"]
                elif metric == "Misses":
                    metric_values[detector_name] = results[dataset_name][detector_name]["misses"]
                elif metric == "Evaluation Time":
                    metric_values[detector_name] = results[dataset_name][detector_name]["time"]

            # Plot lines for each detector
            ax = axes[i // 2, i % 2]  # Adjust axes index for 2x2 grid
            for detector_name in detector_names:
                ax.plot(time_points, metric_values[detector_name], label=detector_name, marker='o')

            ax.set_title(metric)
            ax.set_xlabel("Time (Chunks)")
            ax.set_ylabel(metric)
            ax.legend()

        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()    

# Visualize the results over time
visualize_results_over_time(final_results_NB)
```

## Part 2: Ensemble learning methods 

Sine, Hyperplane, RandomRBF + AccuracyWeightedEnsembleClassifier, DynamicWeightedMajorityClassifier

```python
from skmultiflow.data import SineGenerator, HyperplaneGenerator, RandomRBFGenerator
from skmultiflow.meta import AccuracyWeightedEnsembleClassifier, DynamicWeightedMajorityClassifier, LearnPPNSEClassifier
from skmultiflow.evaluation import EvaluatePrequential

# Define datasets with varying features
datasets = {
    "SineGenerator (Abrupt drift with a smooth, periodic change)": SineGenerator(has_noise=False, random_state=42),
    # Adjusting the rest to have 2 features & remove MIXED
    "HyperplaneGenerator (Gradual drift with a simple linear decision boundary)": HyperplaneGenerator(noise_percentage=0.1, n_drift_features=2, random_state=42, n_features=2),
    #"MIXEDGenerator (Gradual drift with noise and multiple features)": MIXEDGenerator(random_state=42, n_features=2),
    "RandomRBFGenerator (Gradual drift with nonlinear transitions)": RandomRBFGenerator(model_random_state=42, sample_random_state=42, n_features=2)
}

# Instantiate the ensemble classifiers
accuracy_weighted_ensemble = AccuracyWeightedEnsembleClassifier(n_estimators=5)
dynamic_weighted_majority = DynamicWeightedMajorityClassifier(n_estimators=5)
learnpp_nse_classifier = LearnPPNSEClassifier(n_estimators=5) #base_estimator:default=DecisionTreeClassifier

# Set up the evaluator with additional metrics
evaluator = EvaluatePrequential(show_plot=False, metrics=['accuracy', 'precision', 'recall', 'f1'])

# Evaluate each dataset
for dataset_name, stream in datasets.items():
    print(f"Evaluating on dataset: {dataset_name}")
    print()
    
    # Reset the stream to the beginning for each evaluation
    stream.restart()
    
    # Evaluate AccuracyWeightedEnsembleClassifier
    print("1. Evaluating AccuracyWeightedEnsembleClassifier...")
    evaluator.evaluate(stream=stream, model=accuracy_weighted_ensemble)
    print()
    
    # Reset stream and reinitialize model for the next classifier
    stream.restart()
    
    # Evaluate DynamicWeightedMajorityClassifier
    print("2. Evaluating DynamicWeightedMajorityClassifier...")
    evaluator.evaluate(stream=stream, model=dynamic_weighted_majority)
    print()
    
    # Reset stream and reinitialize model for the next classifier
    stream.restart()
    
    # Evaluate LearnPPNSEClassifier
    print("3. Evaluating LearnPPNSEClassifier...")
    evaluator.evaluate(stream=stream, model=learnpp_nse_classifier)
    print()
    
    print()
```
