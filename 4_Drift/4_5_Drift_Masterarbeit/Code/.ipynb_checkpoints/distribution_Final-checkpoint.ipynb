## Step 1: Define functions

```python
# import stats
from evidently.calculations.stattests import hellinger_stat_test, kl_div_stat_test, ks_stat_test, wasserstein_stat_test #Verteilungsdissimilarität
from evidently.calculations.stattests import t_test                                                                     #Distanzmaße zwischen Beobachtungen
```

| Algorithm              | tested Threshold | 
|------------------------|---------------------|
| Hellinger Distance     | 0.1                 | 
| KL Divergence          | 0.15                | 
| Kolmogorov-Smirnov     | 0.05                | 
| Wasserstein Distance   | 0.1                 | 
| t_test     | 0.05                 |

```python
# define functions

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

def plot_example(ref: pd.Series, curr: pd.Series):
  fig = plt.figure(constrained_layout=True, figsize=(15,7))

  gs = GridSpec(2, 3, figure=fig)
  ax1 = fig.add_subplot(gs[0, :])
  ax2 = fig.add_subplot(gs[1, 0])
  ax3 = fig.add_subplot(gs[1, 1])
  ax4 = fig.add_subplot(gs[1, 2])

  # plot feature in time
  ref_points = int(np.round(150 * len(ref) /(len(ref) + len(curr))))
  curr_points = 150 - ref_points

  ref_in_time = [np.mean(x) for x in np.array_split(ref, ref_points)]
  curr_in_time = [np.mean(x) for x in np.array_split(curr, curr_points)]

  ax1.plot(range(ref_points), ref_in_time) #, color=GREY
  ax1.plot(range(ref_points, ref_points + curr_points), curr_in_time) #, color=RED

  # plot referense distr
  sns.histplot(ref, ax=ax2) #color=GREY, 
  # plot current distr
  sns.histplot(curr, ax=ax3) #color=RED,
  # plot two distr
  sns.histplot(ref, ax=ax4) #color=GREY,
  sns.histplot(curr, ax=ax4) #color=RED,
  plt.show()

def evaluare_drift(ref: pd.Series, curr: pd.Series):
  report = pd.DataFrame(columns=['stat_test', 'drift_score', 'is_drifted'])
  for i, (stattest, threshold) in enumerate(
      zip(
          [hellinger_stat_test, kl_div_stat_test, ks_stat_test, wasserstein_stat_test, t_test],
          [0.1, 0.15, 0.05, 0.1, 0.05]
      )
  ):
    report.loc[i, 'stat_test'] = stattest.display_name
    report.loc[i, 'drift_score'], report.loc[i, 'is_drifted'] = stattest.func(ref, curr, 'num', threshold)
  return report
```

## Step 2: Split data

```python
from sklearn.datasets import fetch_california_housing
df = fetch_california_housing(as_frame=True).frame
df.head()
```

```python
# Split
ref = df.MedInc[:10000]
curr = df.MedInc[10000:]
```

```python
# get resuts
plot_example(ref, curr)
#evaluare_drift(ref, curr)
```

## Step 3: Add drift

Shift the whole distribution for each feature by a fixed value: **(alpha + mean(feature)) * perc**

```python
def add_drift(curr: pd.Series, drift_size: float, drift_ratio: float, drift_mode: str='fixed') -> pd.Series:
  alpha = 0.001
  i = int(np.round(len(curr) * drift_ratio))
  if drift_mode == 'fixed':
    delta = (alpha + np.mean(curr)) * drift_size
    curr[:i] = curr[:i] + delta
  else:
    curr[:i] = curr[:i]*(1 + drift_size)
  return curr
```

## Step 4: Test with different drift_size and drift_ratio

```python
# Define ranges for drift_size and drift_ratio
drift_sizes = [0.05, 0.1, 0.3, 0.7, 1] # can the test react to the drift in one dataset segment? 
drift_ratios = [0.01, 0.05, 0.1, 0.2, 0.5] #can the test detect a "small" change in the data, or only a "large" one?

# Iterate through each combination of drift_size and drift_ratio
for drift_size in drift_sizes:
    for drift_ratio in drift_ratios:
        print(f"\033[1mTesting with {drift_size*100}% segment moved by {drift_ratio*100}%")
        
        # Apply drift
        ref = df.MedInc[:10000]
        curr = df.MedInc[10000:]
        curr_shifted = add_drift(curr, drift_size=drift_size, drift_ratio=drift_ratio, drift_mode='fixed')
        
        # Plot and evaluate
        plot_example(ref, curr_shifted) #see plots in another version
        print(f"Results for {drift_size*100}% segment moved by {drift_ratio*100}%")
        print(evaluare_drift(ref, curr_shifted))
        print()
```
